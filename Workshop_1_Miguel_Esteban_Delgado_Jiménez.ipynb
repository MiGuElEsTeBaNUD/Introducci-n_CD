### Problem 1.1

Find a list of all of all of the names in the following string using _regex_.

````python
import re
def names():
    """
    Find all names in a chain according to pattern in a regular expression

    Args:
        simple_string (str): chain to search names
        pattern (str): regular expresion result of analisis

    Return: 
        List of names founds in the chain.
    """
        
    simple_string = """Amy is 5 years old, and her sister Mary is 2 years old. 
    Ruth and Peter, their parents, have 3 kids."""

    pattern = r'\b[A-Z][a-z]*\b'
    
    return re.findall(pattern, simple_string) 
names()
````

### Problem 1.2

The _dataset file_ in [assets/grades.txt](./assets/grades.txt) contains a line separated _list of people_ with their __grade__ in a class. Create a _regex_ to generate a list of just those students who received a __B__ in the course.

````python
import re
"""

This algorythm create a regex to generate a list of just those students who received a B in the course.

Args:
    grades: filename(grades.txt)

Return: 
    A list of students who received a "B"
"""

def grades():
    with open("WORKSHOP_1/txt/grades.txt", "r") as file:
        grades = file.read()

    pattern = re.compile(r'^\s*(.+?)\s*:\s*B\s*$', re.MULTILINE)
    matches = pattern.findall(grades)
    students = matches

    return students
grades()
````

````python
# example of test case
assert len(grades()) == 16
````

````python
import re

"""
 Read the log file and extract the relevant information using regular expressions.

 Returns:
 list: A list of dictionaries, each containing record information.
 """

def logs():
    with open("WORKSHOP_1/txt/logdata.txt", "r") as file:
        logdata = file.read()
    actions = re.split(r"\n", logdata)

    logs_list = []

    # YOUR CODE HERE
    regex = re.compile(
    r'(?P<host>(?:\d+\.){3}\d+)\s+'
    r'(?:\S+)\s+'
    r'(?P<user_name>\S+)\s+'
    r'\[(?P<time>[-+\w\s:/]+)\]\s+'
    r'"(?P<request>POST.+?)"'
)
    for m in regex.finditer(logdata):
        logs_list.append(m.groupdict())
    return logs_list
logs()
````

# Problem 2.1

Write a function called _proportion\_of\_education_ which returns the proportion of __children__ in the dataset who had a mother with the education levels equal to less than high school ($<12$), high school ($12$), more than high school but not a college graduate ($>12$) and _college degree_.

This _function_ should return a __dictionary__ in the form of (use the _correct numbers_, do not round numbers):

```python
{
    "less than high school": 0.2,
    "high school": 0.4,
    "more than high school but not college": 0.2,
    "college": 0.2
}
```

````python
import pandas as pd
import numpy as np

"""
 Calculates the proportion of different educational levels in the loaded DataFrame.

 Returns:
 dict: A dictionary with the proportions of each educational level.

"""

df = pd.read_csv("WORKSHOP_1/csv/NISPUF17.csv", index_col=0)
def proportion_of_education():
    proportions = {}

    EDUS = df['EDUC1']
    edus = np.sort(EDUS.values)

    proportions = {
        "less than high school": 0,
        "high school": 0,
        "more than high school but not college": 0,
        "college": 0}
    n = len(edus)

    proportions["less than high school"] = np.sum(edus == 1)/n
    proportions["high school"] = np.sum(edus == 2)/n
    proportions["more than high school but not college"] = np.sum(edus == 3)/n
    proportions["college"] = np.sum(edus == 4)/n

    return proportions
proportion_of_education()
````
